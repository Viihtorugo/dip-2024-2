```python
import cv2 as cv
import os
import matplotlib.pyplot as plt
```

```python
def create_array_hist(data):
    hist = [0] * 256
    for row in data:
        for pixel in row:
            hist[pixel] += 1
    return hist
```

```python
def plot_hist(path):
    img = cv.imread(path)
    r, g, b = img[:, :, 0], img[:, :, 1], img[:, :, 2]

    hist_r = create_array_hist(r)
    hist_g = create_array_hist(g)
    hist_b = create_array_hist(b)

    plt.figure(figsize=(10, 4))
    plt.plot(hist_r, color='red', label='Red')
    plt.plot(hist_g, color='green', label='Green')
    plt.plot(hist_b, color='blue', label='Blue')
    plt.title(path)
    plt.xlabel('Intensidade')
    plt.ylabel('FrequÃªncia')
    plt.legend()
    plt.grid(False)
    plt.tight_layout()
    plt.show()
```

```python
files = os.listdir("imgs/")

files = [f for f in files if os.path.isfile(os.path.join("imgs/", f))]
```

```python
for i in range(len(files)):
    path = "imgs/" + files[i]
    plot_hist(path)
```
